```python
# Opção para concatenar rápido WoS files no formato Excel
wos = pd.read_excel('savedrecs.xls', 'savedrecs')
wos1 = pd.read_excel('savedrecs (1).xls', 'savedrecs')
wos2 = pd.read_excel('savedrecs (2).xls', 'savedrecs')
wos3 = pd.read_excel('savedrecs (3).xls', 'savedrecs')
wos4 = pd.read_excel('savedrecs (4).xls', 'savedrecs')
wos5 = pd.read_excel('savedrecs (5).xls', 'savedrecs')
```

```python
frames = [wos,wos1,wos2,wos3,wos4,wos5] #,wos6,wos7,wos8,wos9,wos10,wos11]
result = pd.DataFrame().append(frames)
#result = pd.concat(frames)
result
```

```python
# dropping duplicate values
print(len(result))
result.drop_duplicates(keep=False,inplace=True)
print(len(result))
```

```python
result.to_csv("savedrecs_base.csv", index=False)
```
